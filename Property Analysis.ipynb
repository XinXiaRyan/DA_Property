# Data Preprocessing

## Clean mal-formatted data

```python
df[['Day','Month','Year']] = df['Date'].str.split("/",expand= True) #split Date to day,month,year
df['Lattitude'],df['Longtitude']=df.Lattitude.round(4),df.Longtitude.round(4)#round to 4 decimal place.
```

## Fill na data

```python
df.isnull().sum().sort_values(ascending=False) #find na
```

```python
#fill CouncilArea
for i in df.loc[df['CouncilArea'].isna(),'Suburb'].unique():
    df.loc[df['Suburb'].str.contains(i),'CouncilArea']=df.loc[df['Suburb']==i]['CouncilArea'].unique()[0]
```

```python
#fill na
df['Car'].fillna(0, inplace = True)
df['YearBuilt'].fillna('Unknown', inplace = True)
df['BuildingArea'].fillna('Unknown', inplace = True)
```

```python
df.isnull().sum().sort_values(ascending=False) #find na
```
